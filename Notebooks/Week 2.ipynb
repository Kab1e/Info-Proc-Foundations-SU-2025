```python
## Conditional statement (if-while) in a function.
def is_prime(n):
    if n in [2, 3]:
        return True
    if (n == 1) or (n % 2 == 0):
        return False
    r = 3
    while r * r <= n:
        if n % r == 0:
            return False
        r += 2
    return True
print("78 is", is_prime(78))
print("79 is", is_prime(79))
print("80 is", is_prime(80))
```

```python

## A simple Python function to check whether x is even or odd.
def evenOdd(x):
    if (x % 2 == 0):
        print("even")
    else:
        print("odd")


## Driver code to call the function
## evenOdd(2)
evenOdd(3)
```

```python
## Python program to demonstrate default arguments.
## A default argument is a parameter that assumes a default value
## if a value is not provided in the function call for that argument.
def myFun(x, y = 50):
    print("x: ", x)
    print("y: ", y)


## Driver code: We call myFun() with only argument.
myFun(10)
```

```python
## Python program to illustrate
## *argv = variable number of arguments
def myFun(*argv):
    for arg in argv:
        print(arg)

## Driver code
myFun('Welcome', 'to', 'IS 430')
```

```python
## Python program to illustrate
## *kwargs = variable number of keyword arguments

def myFun(**kwargs):
    for key, value in kwargs.items():
        print("%s == %s" % (key, value))

## Driver code
myFun(first='Welcome', last='IS 430', mid='to')
```

```python
## Function for creating a modified list. Define the function pure_func, then define variables in subsequent script.
def pure_func(List):

    New_List = []

    for i in List:
        New_List.append(i**2)

    return New_List

## for i (index value) in Original_List, append list with i^2. (e.g. 1**2 is 1, 2**2 is 4, 3**3 is 9). Return Modified List.

## Specification and display. Define Original_List and Modified_List (as the output of for loop in function pure_func).
Original_List = [1, 2, 3, 4]
Modified_List = pure_func(Original_List)

print("Original List:", Original_List)
print("Modified List:", Modified_List)
```

```python
## Function for recursively adding together all elements in a vector.
def Sum(L, i, n, count):

    # Base case
    if n <= i:
        return count

    count += L[i]

    # Going into the recursion
    count = Sum(L, i + 1, n, count)

    return count

## Specification and display. Define inputs L, i, and n of the function Sum.
L = [1, 2, 3, 4, 5]
count = 0
n = len(L)
print(Sum(L, 0, n, count))
```
